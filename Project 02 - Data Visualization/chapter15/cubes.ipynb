```python
import matplotlib.pyplot as plt
```

```python
# 15-1. Cubes: A number raised to the third power is a cube. Plot the first five
# cubic numbers, and then plot the first 5000 cubic numbers.

x_lst = [1, 2, 3, 4, 5]
y_lst = [1, 8, 27, 64, 125]

plt.scatter(x_lst, y_lst)
plt.title('Cube of Numbers', fontsize=24)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Cube of Value', fontsize=14)
plt.tick_params(axis='both', labelsize=14)

plt.axis([0, 6, 0, 150])

plt.show()
```

```python
x_lst_5k = list(range(1, 5001))
y_lst_5k = [x**3 for x in x_lst_5k]

plt.scatter(x_lst_5k, y_lst_5k, s=30, c=y_lst_5k, cmap=plt.cm.PuBu)
plt.title('Cube of Numbers', fontsize=24)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Cube of Value', fontsize=14)
plt.tick_params(axis='both', labelsize=14)

plt.axis([0, 5500, 0, 130000000000])

plt.show()
```

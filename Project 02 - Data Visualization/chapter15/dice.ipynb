```python
from random import randint
import pygal
```

```python
class Die():
    """A class that represents a single die."""
    
    def __init__(self, num_sides=6):
        """It considers a six-side die."""
        self.num_sides = num_sides
        
        
    def roll(self):
        """It returns a value between 1 and the total number of sides."""
        return randint(1, self.num_sides)
        
```

```python
play01 = Die()
play02 = Die()
```

```python
# Make some rolls and store the results in a list.

results = []

for num in range(1000):
    result = play01.roll() * play02.roll()
    results.append(result)
```

```python
# Analyze the results

frequencies = []
max_result = play01.num_sides + play02.num_sides

for value in range(1, max_result+1):
    frequency = results.count(value)
    frequencies.append(frequency)
    
print(frequencies)
```

```python
# Visualize the results
hist = pygal.Bar()
```

```python
x_labels = [x for x in range(2, max_result+1)]

hist.title = "Results of rolling two D6 dice 1000 times."
hist.x_labels = x_labels
hist.x_title = "Result"
hist.y_title = "Frequency of Result"

hist.add('D6 * D6', frequencies)
hist.render_to_file('dice_visual.svg')
```
